# ¿Cuáles son las habilidades más requeridas en el top 3 de roles con datos en US?

## Metodología
1. Limpieza de habilidades por columna
2. Contar habilidades por job_title_short
3. Calcular porcentajes
4. Graficar

```python
# Importar librerias
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from datasets import load_dataset
import ast

# Cargar datos
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Limpieza de datos
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x) #Se transforman los string a la estructura correspondiente (list)
```

```python
df_us = df[df['job_country'] == 'United States']
df_us.head(3)
```

```python
df_skills = df_us.explode('job_skills') #Se separan los valores de las listas en filas distintas
df_skills
```
