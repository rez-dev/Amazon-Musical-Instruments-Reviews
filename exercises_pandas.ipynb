```python
import pandas as pd
from datasets import load_dataset

# Loading data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

## Análisis a trabajos de US

```python
us_jobs = df[df['job_country'] == 'United States']
us_jobs
```

```python
# limpiar nans
us_jobs = us_jobs[us_jobs['salary_year_avg'].notna()]
us_jobs
```

```python
us_jobs['job_title_short'].unique()
```

```python
# agrupa por trabajo y cuenta para cada columna
us_jobs.groupby('job_title_short').count()
```

```python
# us_jobs.head(3)
us_jobs.groupby('job_title_short')['salary_year_avg'].agg(['median', 'min', 'max', 'count']).sort_values(by='median', ascending=False)
```
